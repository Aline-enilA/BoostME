```python
import pandas as pd

df_fitness = pd.read_csv("market_fitness_fr.csv")
df_cuisine = pd.read_csv("market_cuisine_fr.csv")
df_jeux = pd.read_csv("market_jeux_video_fr.csv")

df_fitness["niche"] = "fitness"
df_cuisine["niche"] = "cuisine"
df_jeux["niche"] = "jeux_video"

df = pd.concat([df_fitness, df_cuisine, df_jeux], ignore_index=True)
df.head()
```

```python
df[["niche","views","likes","comments"]].groupby("niche").describe().round(2)
```

```python
df["engagement_rate"] = (df["likes"] + df["comments"]) / df["views"].replace(0, 1)
df.groupby("niche")["engagement_rate"].describe().round(4)
```

```python
df["published_at"] = pd.to_datetime(df["published_at"], errors="coerce", utc=True)
df["weekday"] = df["published_at"].dt.day_name()
df["hour"] = df["published_at"].dt.hour

df.groupby(["niche","weekday"])["views"].median().sort_values(ascending=False).head(15)
```
