```python
import pandas as pd

df_fitness = pd.read_csv("market_fitness_fr.csv")
df_cuisine = pd.read_csv("market_cuisine_fr.csv")
df_jeux = pd.read_csv("market_jeux_video_fr.csv")
```

```python
df_fitness["niche"] = "fitness"
df_cuisine["niche"] = "cuisine"
df_jeux["niche"] = "jeux_video"
```

```python
df_market = pd.concat(
    [df_fitness, df_cuisine, df_jeux],
    ignore_index=True
)

df_market.head()
```

```python
df_market = df_market.drop_duplicates(subset="video_id")
```

```python
num_cols = ["views", "likes", "comments"]

for col in num_cols:
    df_market[col] = pd.to_numeric(df_market[col], errors="coerce").fillna(0).astype(int)

```

```python
df_market = df_market[
    [
        "video_id",
        "niche",
        "title",
        "channel",
        "published_at",
        "views",
        "likes",
        "comments"
    ]
]
```

```python
df_market.to_csv("youtube_market_fr_clean.csv", index=False)
print("youtube_market_fr_clean.csv crÃ©Ã© âœ…")
```
